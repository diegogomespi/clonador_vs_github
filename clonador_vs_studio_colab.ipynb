```python
#@title INICIAR CLONADOR VS STUDIO DO GITHUB
REPO_URL = "https://github.com/diegogomespi/clonador_vs_github.git"
BRANCH = "main"
PROJECT_DIR = "/content/clonador_vs_github"

# Opcao antiga: Telegram direto pelo Colab.
TELEGRAM_BOT_TOKEN = ""
TELEGRAM_CHAT_ID = ""

# Opcao recomendada: Cloudflare Worker escondendo o token.
TELEGRAM_PROXY_URL = ""
TELEGRAM_PROXY_SECRET = ""
TELEGRAM_SEND_AUDIO = True
TELEGRAM_SILENT = False

import os
import subprocess
import sys

parent_dir = os.path.dirname(PROJECT_DIR)
os.makedirs(parent_dir, exist_ok=True)

if not os.path.exists(PROJECT_DIR):
    subprocess.check_call(["git", "clone", REPO_URL, PROJECT_DIR])

if PROJECT_DIR not in sys.path:
    sys.path.insert(0, PROJECT_DIR)

from colab_launcher import run_from_colab

run_from_colab(
    REPO_URL,
    branch=BRANCH,
    project_dir=PROJECT_DIR,
    telegram_bot_token=TELEGRAM_BOT_TOKEN,
    telegram_chat_id=TELEGRAM_CHAT_ID,
    telegram_send_audio=TELEGRAM_SEND_AUDIO,
    telegram_silent=TELEGRAM_SILENT,
    telegram_proxy_url=TELEGRAM_PROXY_URL,
    telegram_proxy_secret=TELEGRAM_PROXY_SECRET,
)
```
